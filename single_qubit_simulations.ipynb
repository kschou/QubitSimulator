# Setup code

## Pulse types

```python
# define the drive envelope
def gaussian_pulse(t, args):
    '''
        We define a Gaussian pulse normalized to <a> pi rotations with width <sigma> centered around <t0>
    '''
    a = args['A']
    t0 = args['t0']
    sigma = args['sigma']
    scaling_pi = 0.5 * np.pi
    return (a * scaling_pi / (sigma * np.sqrt(2.0 * np.pi)) * np.exp(-(t - t0)**2 / (2.0 * sigma**2)))
gaussian_drive = np.vectorize(gaussian_pulse)
```

```python
# define the drive coefficient
def const_drive_env(t, args):
    a = args['A']
    if t > 0: return a
    else: return 0.0
constant_drive = np.vectorize(const_drive_env)
```

## Setup Hamiltonian

```python
N = 5

a = qt.destroy(N)
g_state = qt.fock(N, 0)
e_state = qt.fock(N, 1)
f_state = qt.fock(N, 2)

# Define time independent Hamiltonian
def get_H_ti(detuning, alpha):
    return -detuning * a.dag() * a - alpha / 2.0 * a.dag() * a.dag() * a * a
H_drive = a + a.dag()

# define collapse operators: relaxation and dephasing
sx = g_state * e_state.dag() + e_state * g_state.dag()
sz = g_state * g_state.dag() - e_state * e_state.dag()
def get_c_ops(kappa, dephasing):
    return [np.sqrt(kappa) * a, np.sqrt(dephasing) *  sz]

# we care primarily about the first three levels of the transmon
pop_projs = [g_state * g_state.dag(), e_state * e_state.dag(), f_state * f_state.dag()] 
```

## Rabi experiments

```python
def power_rabi_expt(amps, sigma, t0=0.0, H_params=(0.0, 100.0), expects=pop_projs, c_ops=[]):
    
    final_expects = []
    for a in amps:
        args = {'A': a, 't0': t0, 'sigma': sigma}
        
        detuning, alpha = H_params
        expt = qs.Experiment(get_H_ti(detuning, alpha))
        expt.setup_experiment(g_state, c_ops=c_ops, expects=expects)
        
        time_range = (args['t0'] + args['sigma']*5, args['t0'] - args['sigma']*5, 101)
        
        expt.define_time_dependence([
            [(H_drive, gaussian_pulse, args), time_range]
            ])
        
        expt.simulate()
        expt.calculate_expects()
        
        final_expects.append(expt.sim_expects[:, -1])
        
    return np.array(final_expects)

def time_rabi_expt(sigmas, base_amp=1.0, t0=0.0, H_params=(0.0, 100.0), expects=pop_projs, c_ops=[]):
    
    final_expects = []
    for s in sigmas:
        
        a = base_amp * s
        args = {'A': a, 't0': t0, 'sigma': s}
        
        detuning, alpha = H_params
        expt = qs.Experiment(get_H_ti(detuning, alpha))
        expt.setup_experiment(g_state, c_ops=c_ops, expects=expects)
        
        time_range = (args['t0'] + args['sigma']*5, args['t0'] - args['sigma']*5, 101)
        
        expt.define_time_dependence([
            [(H_drive, gaussian_pulse, args), time_range]
            ])
        
        expt.simulate()
        expt.calculate_expects()
        
        final_expects.append(expt.sim_expects[:, -1])
        
    return np.array(final_expects)

def determine_final_state(expects):
    '''
        fock number, amp, time
    '''
    
    return expects[:, :, -1]
```

# 2. With decoherence: $\kappa$ and dephasing

## Decay during qubit pulse

```python
detuning, alpha = (0.0, 200.0)
kappa, dephasing = (1/100.0, 0.0)

sigmas = np.linspace(1.0, 0.1, 11)

pd_tlists = []
pd_expects = []
for s in sigmas:
    args = {'A': 1, 't0': 0.0, 'sigma': s}
    
    pidecay = qs.Experiment(get_H_ti(detuning, alpha))
    pidecay.setup_experiment(g_state, c_ops=get_c_ops(kappa, dephasing), expects=pop_projs)
    
    pi_time_range = (args['t0'] - args['sigma']*5, args['t0'] + args['sigma']*5, 101)
    
    pidecay.define_time_dependence([
        [(H_drive, gaussian_pulse, args), pi_time_range]
        ])
    
    pidecay.simulate()
    pidecay.calculate_expects()
    
    pd_tlists.append(pidecay.sim_tlist)
    pd_expects.append(pidecay.sim_expects)
    
pd_tlists = np.array(pd_tlists)
pd_expects = np.array(pd_expects)
```

```python
pd_final_pops = np.array([ expects[:,-1] for expects in pd_expects])
pd_pulse_len = np.array([ (tlist[-1] - tlist[0]) for tlist in pd_tlists ])

exp_decay = np.exp(- pd_pulse_len *kappa)
```

```python
fig = plt.figure(figsize=(12, 4))

ax = fig.add_subplot(121)

ax.plot(sigmas, pd_final_pops[:, 0], 'rs-')
ax.plot(sigmas, pd_final_pops[:, 1], 'bs-')
ax.plot(sigmas, pd_final_pops[:, 2], 'gs-')
ax.legend( ('g', 'e', 'f') )
ax.set_xlabel('$\pi$ pulse sigma')
ax.set_ylabel('population')


ax = fig.add_subplot(122)
ax.plot(sigmas, pd_final_pops[:,1], 'bs-', label='e')
ax.plot(sigmas, exp_decay, 'k-', label='$T_1$ decay')
```

### $T_1$ decay

```python
detuning, alpha = (0.0, 200.0)
kappa, dephasing = (1/10.0, 0.0)
args = {'A': 1, 't0': 0.0, 'sigma': 0.2}

t1decay = qs.Experiment(get_H_ti(detuning, alpha))
t1decay.setup_experiment(g_state, c_ops=get_c_ops(kappa, dephasing), expects=pop_projs)

pi_time_range = (args['t0'] - args['sigma']*5, args['t0'] + args['sigma']*5, 101)
decay_time_range = (args['t0'] + args['sigma']*5, args['t0'] + args['sigma']*5 + 3 / kappa, 51) 
end_pulse_time = args['t0'] + args['sigma']*5

t1decay.define_time_dependence([
    [(H_drive, gaussian_pulse, args), pi_time_range],
    [None, decay_time_range]
    ])

t1decay.simulate()
t1decay.calculate_expects()

times = t1decay.sim_tlist
t1expects = t1decay.sim_expects
```

```python
decay_start = np.where(times == end_pulse_time)[0][0]

decay_times = times[decay_start:]
decay_expects = t1expects[:,decay_start:]

fit = fp.Exp(x_data=decay_times, y_data=decay_expects[1])
lm_min, params, y_final = fit.fit(plot=False)
fp.report_fit_params(params)
```

```python
fig = plt.figure(figsize=(10, 4))

ax = fig.add_subplot(121)
ax.plot(times, t1expects[0], 'r-')
ax.plot(times, t1expects[1], 'b-')
ax.plot(times, t1expects[2], 'g-')
ax.legend( ('g', 'e', 'f') )
ax.set_xlabel('time')
ax.set_ylabel('population')
ax.axvline(x=end_pulse_time, color='k')

ax = fig.add_subplot(122)
ax.plot(decay_times, decay_expects[1], 'b.-')
ax.plot(decay_times, y_final, 'k-')
ax.set_yscale('log')
```
